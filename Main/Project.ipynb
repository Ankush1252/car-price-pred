# Problem Statement

        -> Our objective is to predict the price of car

## Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler,OneHotEncoder,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.feature_selection import SequentialFeatureSelector

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

# Data Gathering

```python
def DataLoader(path):
    data1 = pd.read_csv(path)
    return data1
```

```python
ds = DataLoader('../data/Cars93.csv')
ds
```

# Define X and Y variable

```python
x=ds.drop(['Price','id'],axis=1)             #independent Feature
y=ds['Price'] 
```

# Preprocessing

```python
con=x.select_dtypes(exclude='object').columns       #list of numerical variable
cat=x.select_dtypes(include='object').columns       # List of Categorical variable

num_pipe=Pipeline(steps=[('Imputer',SimpleImputer(strategy='median')),('Scalar',StandardScaler())])
cat_pipe=Pipeline(steps=[('Imputer',SimpleImputer(strategy='most_frequent')),('encoder',OrdinalEncoder())])

preprocessor=ColumnTransformer([('Num_pipe',num_pipe,con),('cat_pipe',cat_pipe,cat)])

x1=pd.DataFrame(preprocessor.fit_transform(x),columns=preprocessor.get_feature_names_out())

x1
```

```python
x1
```

# Feature Selection

```python
lr=LinearRegression()
sfs=SequentialFeatureSelector(lr,n_jobs=-1)
x2=pd.DataFrame(sfs.fit_transform(x1,y),columns=sfs.get_feature_names_out())
x2
```

## Splitting Data Into Train and Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(x2,y,test_size=0.2,random_state=2)
```

# Model

```python
linear_reg=LinearRegression()
linear_reg.fit(x_train,y_train)
```

# Evaluation 

### Training Data Evaluation 

```python
tr_pred=linear_reg.predict(x_train)

mse=mean_squared_error(tr_pred,y_train)
rmse=np.sqrt(mse)
mae=mean_absolute_error(tr_pred,y_train)
r2=r2_score(tr_pred,y_train)

print('MSE:',mse)
print('RMSE:',rmse)
print('MAE:',mae)
print('R2 SCore:',r2)
```

### Testing Data Evaluation

```python
ts_pred=linear_reg.predict(x_test)

mse1=mean_squared_error(ts_pred,y_test)
rmse1=np.sqrt(mse1)
mae1=mean_absolute_error(ts_pred,y_test)
r1=r2_score(ts_pred,y_test)

print('MSE:',mse1)
print('RMSE:',rmse1)
print('MAE:',mae1)
print('R2 SCore:',r1)
```
